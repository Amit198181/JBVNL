```python
# The manual mapping created in the previous step was incorrect because the index of df_transposed was
# already in the 'FY XXXX-XX' format but with incorrect year numbers.
# I need to create a mapping from the incorrect 'FY' index values in df_transposed
# to the correct 'FY' index values in df_load based on their order.

# Get the incorrect FY index values from df_transposed
incorrect_fy_index = df_transposed.index.tolist()

# Get the correct FY index values from df_load
correct_fy_index = df_load.index.tolist()

# Create a mapping dictionary by zipping the incorrect and correct FY index values
correct_fy_mapping = dict(zip(incorrect_fy_index, correct_fy_index))

# Rename the index of df_transposed using the correct FY mapping
df_transposed_final_index = df_transposed.rename(index=correct_fy_mapping)

# Align the two DataFrames based on the corrected financial year index
aligned_df = pd.merge(df_transposed_final_index, df_load, left_index=True, right_index=True, how='inner')

# Display the head of the aligned DataFrame to confirm the merge was successful
print("\nAligned DataFrame head after final index mapping and merging:")
display(aligned_df.head())

# Now that the dataframes are aligned, proceed with the sensitivity analysis steps.

# 1. Calculate the total number of consumers for each financial year in the aligned_df
# Sum across all columns except 'Connected Load (kVA)'
aligned_df['Total Consumers'] = aligned_df.drop(columns=['Connected Load (kVA)']).sum(axis=1)

# 2. Calculate the ratio of 'Connected Load (kVA)' to 'Total Consumers' for each financial year
aligned_df['Load per Consumer Ratio'] = aligned_df['Connected Load (kVA)'] / aligned_df['Total Consumers']

# 3. Select a base financial year for the sensitivity analysis
# Using the last financial year with complete data as the base year
base_year = aligned_df.index[-1]
print(f"\nBase financial year for sensitivity analysis: {base_year}")

# 4. For the base financial year, calculate the percentage of total consumers for each category.
base_year_data = aligned_df.loc[base_year]
consumer_categories = aligned_df.columns.drop(['Connected Load (kVA)', 'Total Consumers', 'Load per Consumer Ratio'])
category_consumer_counts_base = base_year_data[consumer_categories]
total_consumers_base = base_year_data['Total Consumers']
category_percentage_base = (category_consumer_counts_base / total_consumers_base) * 100

print(f"\nPercentage of total consumers for each category in {base_year}:")
display(category_percentage_base)

# 5. To perform a basic sensitivity analysis, simulate a percentage increase (e.g., 10%) in the number of consumers for a specific category
# 6. Calculate the new total number of consumers after the simulated increase.
# 7. Using the 'Load per Consumer Ratio' from the base year, estimate the new 'Connected Load (kVA)' based on the new total number of consumers.
# 8. Calculate the percentage change in the 'Connected Load (kVA)' resulting from the simulated increase in the specific category's consumer number.
# 9. Repeat steps 5-8 for each consumer category to understand the impact of a percentage increase in each category on the total connected load.
# 10. Store the results of the sensitivity analysis (category, simulated increase percentage, resulting load percentage change) in a suitable format (e.g., a new DataFrame).

sensitivity_results = []
simulated_increase_percentage = 0.10 # 10% increase

base_load_per_consumer_ratio = base_year_data['Load per Consumer Ratio']
base_connected_load = base_year_data['Connected Load (kVA)']

for category in consumer_categories:
    # Simulate increase in the current category's consumer number for the base year
    simulated_consumer_counts_base = category_consumer_counts_base.copy()
    simulated_consumer_counts_base[category] *= (1 + simulated_increase_percentage)

    # Calculate the new total number of consumers
    new_total_consumers = simulated_consumer_counts_base.sum()

    # Estimate the new 'Connected Load (kVA)' using the base year's load per consumer ratio
    estimated_new_connected_load = new_total_consumers * base_load_per_consumer_ratio

    # Calculate the percentage change in the 'Connected Load (kVA)'
    percentage_change_in_load = ((estimated_new_connected_load - base_connected_load) / base_connected_load) * 100

    # Store the results
    sensitivity_results.append({
        'Category': category,
        'Simulated Increase (%)': simulated_increase_percentage * 100,
        'Resulting Load Change (%)': percentage_change_in_load
    })

# Convert the results to a DataFrame
sensitivity_df = pd.DataFrame(sensitivity_results)

print("\nSensitivity Analysis Results (10% increase in each category):")
display(sensitivity_df)
```
