```python
import pandas as pd
from google.colab import files
import io

print("Please upload your Excel file:")
uploaded = files.upload()
```

```python
# Assuming only one file was uploaded
for filename in uploaded.keys():
    excel_file = io.BytesIO(uploaded[filename])
    df = pd.read_excel(excel_file)
    print(f"Successfully loaded '{filename}' into a pandas DataFrame.")
    display(df.head())
```

```python
print(f"DataFrame shape before removing duplicates: {df.shape}")
df.drop_duplicates(subset=['ACCOUNT_NO'], inplace=True)
print(f"DataFrame shape after removing duplicates by 'ACCOUNT_NO': {df.shape}")
```

```python
display(df.describe())
```

```python
# Display all column names to identify relevant MDI and Billed Units columns
print(df.columns.tolist())
```

### Maximum Potential Units Consumed (Based on Contracted Load)

```python
# Calculate the maximum CONTRACTED_LOAD in the entire dataset
max_contracted_load = df['CONTRACTED_LOAD'].max()
print(f"The maximum contracted load among all consumers is: {max_contracted_load}")

# Display a table of all consumers with their CONTRACTED_LOAD
print("\nTable of all consumers and their contracted load:")
display(df[['ACCOUNT_NO', 'CONTRACTED_LOAD']].head())
```

```python
# Assuming February 2026, which has 28 days
total_days_in_month = 28

# Calculate the maximum potential units for each consumer
df['MAX_POTENTIAL_UNITS'] = df['CONTRACTED_LOAD'] * 24 * total_days_in_month

print("Table of consumers with their Contracted Load and Maximum Potential Units:")
display(df[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MAX_POTENTIAL_UNITS']].head())
```

### Comparison of Actual Billed Units with Maximum Potential Units

```python
df['UTILIZATION_RATIO'] = (df['BILLED_UNITS_HTS_202602'] / df['MAX_POTENTIAL_UNITS']) * 100
df['UNDER_UTILIZATION_AMOUNT'] = df['MAX_POTENTIAL_UNITS'] - df['BILLED_UNITS_HTS_202602']

print("Consumers with their Contracted Load, Max Potential Units, Actual Billed Units, Utilization Ratio, and Under-Utilization Amount:")
display(df[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MAX_POTENTIAL_UNITS', 'BILLED_UNITS_HTS_202602', 'UTILIZATION_RATIO', 'UNDER_UTILIZATION_AMOUNT']].head())
```

```python
anomalous_consumers = df[df['BILLED_UNITS_HTS_202602'] > df['MAX_POTENTIAL_UNITS']]

if not anomalous_consumers.empty:
    print(f"Found {len(anomalous_consumers)} consumer(s) where BILLED_UNITS_HTS_202602 exceeds MAX_POTENTIAL_UNITS:")
    display(anomalous_consumers[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MAX_POTENTIAL_UNITS', 'BILLED_UNITS_HTS_202602', 'UTILIZATION_RATIO']].sort_values(by='UTILIZATION_RATIO', ascending=False))
else:
    print("No consumers found where BILLED_UNITS_HTS_202602 exceeds MAX_POTENTIAL_UNITS.")
```

```python
output_filename = 'overdrawn_power_consumers.csv'
anomalous_consumers.to_csv(output_filename, index=False)
print(f"Successfully saved the consumers with overdrawn power to '{output_filename}'")
```

```python
print("Comparison of MDI_HTS_202602 and CONTRACTED_LOAD for anomalous consumers:")
display(anomalous_consumers[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MDI_HTS_202602']])
```

```python
mdi_less_than_cd_in_anomalous = anomalous_consumers[anomalous_consumers['MDI_HTS_202602'] < anomalous_consumers['CONTRACTED_LOAD']]

if not mdi_less_than_cd_in_anomalous.empty:
    print(f"Found {len(mdi_less_than_cd_in_anomalous)} consumer(s) from the 'overdrawn power' list where MDI_HTS_202602 is less than CONTRACTED_LOAD:")
    display(mdi_less_than_cd_in_anomalous[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MDI_HTS_202602', 'BILLED_UNITS_HTS_202602', 'MAX_POTENTIAL_UNITS']])
else:
    print("No consumers found from the 'overdrawn power' list where MDI_HTS_202602 is less than CONTRACTED_LOAD.")
```

### Consumers with Billed Units Exceeding Maximum Potential Units, but MDI Less Than Contracted Load

This section highlights consumers who, despite having their billed units exceed their maximum potential units (indicating overdrawn power), still recorded a Maximum Demand Indicator (MDI) less than their Contracted Load. This could suggest several scenarios:

*   **Billing Anomalies**: There might be discrepancies in how billed units are calculated or recorded.
*   **Load Management**: Consumers might have managed their peak demand (MDI) to stay within the contracted load, even if their total energy consumption (billed units) was high due over prolonged use at lower than peak rates.
*   **Data Entry Errors**: Potential errors in recording MDI or Contracted Load values.

Further investigation into these specific accounts would be beneficial to understand the root cause of this pattern.

```python
print('Details of consumers where BILLED_UNITS_HTS_202602 exceeds MAX_POTENTIAL_UNITS, but MDI_HTS_202602 is less than CONTRACTED_LOAD:')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(mdi_less_than_cd_in_anomalous[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'MDI_HTS_202602', 'BILLED_UNITS_HTS_202602', 'MAX_POTENTIAL_UNITS']])
```

### Validation of Anomalous Consumers Criteria

```python
# Validate that all entries in `mdi_less_than_cd_in_anomalous` satisfy the conditions
validation_check_billed_units = (mdi_less_than_cd_in_anomalous['BILLED_UNITS_HTS_202602'] > mdi_less_than_cd_in_anomalous['MAX_POTENTIAL_UNITS']).all()
validation_check_mdi = (mdi_less_than_cd_in_anomalous['MDI_HTS_202602'] < mdi_less_than_cd_in_anomalous['CONTRACTED_LOAD']).all()

print(f"All listed consumers have BILLED_UNITS_HTS_202602 > MAX_POTENTIAL_UNITS: {validation_check_billed_units}")
print(f"All listed consumers have MDI_HTS_202602 < CONTRACTED_LOAD: {validation_check_mdi}")

if validation_check_billed_units and validation_check_mdi:
    print("Validation successful: All records in 'mdi_less_than_cd_in_anomalous' meet both criteria.")
else:
    print("Validation failed: Some records in 'mdi_less_than_cd_in_anomalous' do not meet the criteria.")
```

```python
output_filename_anomalies = 'mdi_less_than_contracted_load_anomalies.csv'
mdi_less_than_cd_in_anomalous.to_csv(output_filename_anomalies, index=False)
print(f"Successfully saved the consumers with MDI less than Contracted Load from the 'overdrawn power' list to '{output_filename_anomalies}'")
```
