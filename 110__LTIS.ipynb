# Task
Analyze the uploaded Excel file to determine if the maximum demand exceeded the contract demand by 10% or more at least four times during the last financial year.

## Load data

### Subtask:
Load the data from the uploaded Excel file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the Excel file into a DataFrame.



```python
import pandas as pd

df = pd.read_excel('LTIS_APR_AUG.xlsx')
display(df.head())
```

## Identify relevant columns

### Subtask:
Identify the columns containing maximum demand, contract demand, and date information.


**Reasoning**:
Examine the column names in the DataFrame to identify columns related to maximum demand, contract demand, and date information.



```python
print(df.columns)
```

## Calculate excess demand

### Subtask:
Calculate the percentage by which the maximum demand exceeds the contract demand for each entry.


**Reasoning**:
Calculate the percentage excess demand for each maximum demand column and store the results in new columns.



```python
mdi_columns = [col for col in df.columns if 'MDI_HTS_' in col]
for mdi_col in mdi_columns:
    percentage_excess_col_name = f'PERCENT_EXCESS_{mdi_col}'
    df[percentage_excess_col_name] = ((df[mdi_col] - df['CONTRACTED_LOAD']) / df['CONTRACTED_LOAD']) * 100
display(df.head())
```

## Filter occurrences

### Subtask:
Filter the data to identify instances where the excess demand is 10% or more.


**Reasoning**:
Filter the DataFrame to include only rows where the percentage excess demand is 10% or more in any of the relevant columns.



```python
percentage_excess_cols = [col for col in df.columns if 'PERCENT_EXCESS_MDI_HTS_' in col]
filtered_df = df[
    (df[percentage_excess_cols] >= 10).any(axis=1)
].copy()
display(filtered_df.head())
```

## Count occurrences

### Subtask:
Count how many times the maximum demand exceeded the contract demand by 10% or more.


**Reasoning**:
Identify the percentage excess columns, count the number of times the value is 10 or greater for each row, store the count in a new column 'Excess_Demand_Count', and display the head of the dataframe.



```python
percentage_excess_cols = [col for col in filtered_df.columns if 'PERCENT_EXCESS_MDI_HTS_' in col]
filtered_df['Excess_Demand_Count'] = (filtered_df[percentage_excess_cols] >= 10).sum(axis=1)
display(filtered_df.head())
```

## Determine if condition is met

### Subtask:
Check if the count of occurrences is at least four.


**Reasoning**:
Filter the DataFrame to keep only rows where 'Excess_Demand_Count' is 4 or more, then count the number of rows in this filtered DataFrame and print a message based on the count.



```python
final_filtered_df = filtered_df[filtered_df['Excess_Demand_Count'] >= 4]
count_at_least_four = len(final_filtered_df)

if count_at_least_four > 0:
    print(f"At least one consumer exceeded the contract demand by 10% or more at least four times. Number of such consumers: {count_at_least_four}")
else:
    print("No consumers exceeded the contract demand by 10% or more at least four times.")
```

## Output result

### Subtask:
Display the final filtered data containing consumers who exceeded the contract demand by 10% or more at least four times.


**Reasoning**:
Display the final filtered DataFrame containing consumers who exceeded the contract demand by 10% or more at least four times.



```python
display(final_filtered_df)
```

## Calculate Enhanced Contract Demand

### Subtask:
For each consumer who exceeded the contract demand by 10% or more at least four times, calculate the average of their four highest maximum demand recordings.

**Reasoning**:
Select the MDI columns for the filtered consumers, find the four highest MDI values for each consumer, calculate the average of these four values, store the result in a new column 'Enhanced_Contract_Demand', and display the head of the updated DataFrame.

```python
mdi_columns = [col for col in df.columns if 'MDI_HTS_' in col]

# Convert MDI columns to numeric, coercing errors
for col in mdi_columns:
    final_filtered_df[col] = pd.to_numeric(final_filtered_df[col], errors='coerce')


def calculate_enhanced_demand(row):
    mdi_values = pd.to_numeric(row[mdi_columns], errors='coerce').dropna()
    if len(mdi_values) >= 4:
        return mdi_values.nlargest(4).mean()
    return None

final_filtered_df['Enhanced_Contract_Demand'] = final_filtered_df.apply(calculate_enhanced_demand, axis=1)

display(final_filtered_df[['CONSUMER_NAME', 'CONTRACTED_LOAD', 'Excess_Demand_Count', 'Enhanced_Contract_Demand']].head())
```
