```python
import pandas as pd
from google.colab import files
import io

print("Please upload your Excel file:")
uploaded = files.upload()
```

```python
# Assuming only one file was uploaded
for filename in uploaded.keys():
    excel_file = io.BytesIO(uploaded[filename])
    df = pd.read_excel(excel_file)
    print(f"Successfully loaded '{filename}' into a pandas DataFrame.")
    display(df.head())
```

```python
display(df.describe())
```

```python
# Display all column names to identify relevant MDI and Billed Units columns
print(df.columns.tolist())
```

```python
# Define the MDI and Billed Units columns for the last four months
mdi_cols = ['MDI_HTS_202511', 'MDI_HTS_202512', 'MDI_HTS_202601', 'MDI_HTS_202602']
billed_units_cols = ['BILLED_UNITS_HTS_202511', 'BILLED_UNITS_HTS_202512', 'BILLED_UNITS_HTS_202601', 'BILLED_UNITS_HTS_202602']

# Filter for MDI = 0 for the last four months
meter_failure_mdi_filter = (df[mdi_cols] == 0).all(axis=1)

# Filter for monthly consumption > 100 kWh for the last four months
high_consumption_filter = (df[billed_units_cols] > 100).all(axis=1)

# Combine the filters
filtered_consumers = df[meter_failure_mdi_filter & high_consumption_filter]

# Get the list of Consumer IDs
consumer_ids = filtered_consumers['CONSNO'].tolist()

if consumer_ids:
    print("Consumer IDs matching the criteria (MDI=0 for last 4 months AND Billed Units > 100 kWh for last 4 months):\n")
    for cons_id in consumer_ids:
        print(cons_id)
    print("\nPotential causes for such a pattern (MDI=0 with high consumption) could include:")
    print("1. MDI sensor failure: The MDI meter component might be faulty and consistently reporting zero despite actual consumption.")
    print("2. Meter bypass or tampering: The meter's MDI recording mechanism might have been intentionally bypassed.")
    print("3. Data recording error: An error in the data collection or storage system leading to incorrect MDI readings.")
else:
    print("No consumers found matching the specified criteria.")
```

```python
# Define the MDI and Billed Units columns for the most recent month (February 2026)
mdi_recent_month_col = 'MDI_HTS_202602'
billed_units_recent_month_col = 'BILLED_UNITS_HTS_202602'

# Filter for MDI = 0 for the most recent month
meter_failure_recent_filter = (df[mdi_recent_month_col] == 0)

# Filter for monthly consumption > 100 kWh for the most recent month
high_consumption_recent_filter = (df[billed_units_recent_month_col] > 100)

# Combine the filters
filtered_consumers_recent = df[meter_failure_recent_filter & high_consumption_recent_filter]

# Get the list of Consumer IDs
consumer_ids_recent = filtered_consumers_recent['CONSNO'].tolist()

if consumer_ids_recent:
    print(f"Consumer IDs matching the criteria (MDI=0 for {mdi_recent_month_col} AND Billed Units > 100 kWh for {billed_units_recent_month_col}):\n")
    for cons_id in consumer_ids_recent:
        print(cons_id)
    print("\nPotential causes for such a pattern (MDI=0 with high consumption) could include:")
    print("1. MDI sensor failure: The MDI meter component might be faulty and consistently reporting zero despite actual consumption.")
    print("2. Meter bypass or tampering: The meter's MDI recording mechanism might have been intentionally bypassed.")
    print("3. Data recording error: An error in the data collection or storage system leading to incorrect MDI readings.")
else:
    print("No consumers found matching the specified criteria for the most recent month.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant columns for plotting
plot_df = filtered_consumers_recent[['CONSNO', 'MDI_HTS_202602', 'BILLED_UNITS_HTS_202602', 'CONTRACTED_LOAD']]

# Melt the DataFrame for easier plotting with seaborn
plot_df_melted = plot_df.melt(id_vars='CONSNO', var_name='Metric', value_name='Value')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Metric', y='Value', hue='CONSNO', data=plot_df_melted, palette='viridis')
plt.title('Comparison of Key Metrics for Consumers with MDI=0 and Billed Units > 100 kWh (Feb 2026)')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
