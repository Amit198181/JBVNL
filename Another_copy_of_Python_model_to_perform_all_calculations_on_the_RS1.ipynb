The next step is to handle these outliers. Based on the box plots, many columns show potential outliers. A common approach is to cap the outliers using the Interquartile Range (IQR) method. I will calculate the IQR for each numerical column and cap values outside 1.5 times the IQR below the first quartile or above the third quartile. Finally, I will consider feature scaling for the numerical columns since many ML algorithms are sensitive to the scale of features. StandardScaler is a good choice as it standardizes features by removing the mean and scaling to unit variance, which is suitable for a wide range of algorithms.



```python
# Handle outliers using the IQR method (capping)
for col in numerical_cols_for_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

# Apply feature scaling to numerical columns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numerical_cols_for_outliers] = scaler.fit_transform(df[numerical_cols_for_outliers])

# Display the processed data head
display(df.head())
```

## Model training

### Subtask:
Train a K-means clustering model on the prepared financial data.


**Reasoning**:
Select numerical columns, import KMeans, instantiate and fit the model, add cluster labels to the dataframe, and display the head and cluster counts.



```python
# Select the numerical columns for clustering. Exclude 'Particular' and 'SI.No.'
numerical_cols_for_clustering = df.select_dtypes(include=['number']).columns.tolist()
if 'Particular' in numerical_cols_for_clustering:
    numerical_cols_for_clustering.remove('Particular')
if 'SI.No.' in numerical_cols_for_clustering:
    numerical_cols_for_clustering.remove('SI.No.')

X = df[numerical_cols_for_clustering]

# Import KMeans
from sklearn.cluster import KMeans

# Instantiate KMeans with a reasonable number of clusters (e.g., 4) and set random_state
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Fit the KMeans model to the selected numerical data
kmeans.fit(X)

# Add the cluster labels to the original DataFrame
df['Cluster'] = kmeans.labels_

# Display the head of the DataFrame with the new 'Cluster' column
display(df.head())

# Display the count of data points in each cluster
print("\nCluster counts:")
display(df['Cluster'].value_counts())
```

## Model evaluation

### Subtask:
Evaluate the performance of the K-means clustering model using appropriate metrics.


**Reasoning**:
Import the silhouette score metric and calculate it for the clustering results.



```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score
# Use the original numerical data X and the cluster labels from the 'Cluster' column in df
silhouette_avg = silhouette_score(X, df['Cluster'])

# Print the silhouette score
print(f'The average silhouette score is: {silhouette_avg:.2f}')
```
