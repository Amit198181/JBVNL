```python
class PowerUtilityMetrics:
    """
    Calculates key power utility performance metrics (Billing Efficiency,
    T&D Loss, Collection Efficiency, and AT&C Loss) based on energy and revenue data.
    """

    def __init__(self, input_energy, units_billed, assessment_w_o_subsidy, subsidy, collection):
        """
        Initializes the utility data.

        Args:
            input_energy (float): Total energy input to the system (in units/kWh/MU).
            units_billed (float): Energy units billed to consumers (in units/kWh/MU).
            assessment_w_o_subsidy (float): Total assessment/billing without subsidy (in currency).
            subsidy (float): Subsidy amount received from the government (in currency).
            collection (float): Amount collected from consumers (in currency).
        """
        self.input_energy = input_energy
        self.units_billed = units_billed
        self.assessment_w_o_subsidy = assessment_w_o_subsidy
        self.subsidy = subsidy
        self.collection = collection

        # Derived Financial Totals
        self.total_assessment_w_subsidy = self.assessment_w_o_subsidy + self.subsidy
        self.total_collection_w_subsidy = self.collection + self.subsidy


    def calculate_billing_efficiency(self):
        """Calculates Billing Efficiency."""
        if self.input_energy == 0:
            return 0.0
        # Formula: (Units Billed / Input Energy) * 100
        self.billing_efficiency = (self.units_billed / self.input_energy) * 100
        return self.billing_efficiency

    def calculate_td_loss(self):
        """Calculates T&D Loss as per the provided definition."""
        # Formula: (1 - Billing Efficiency/100) * 100
        be_factor = self.calculate_billing_efficiency() / 100
        self.td_loss = (1 - be_factor) * 100
        return self.td_loss

    def calculate_collection_efficiency(self):
        """
        Calculates Collection Efficiency.
        Using the conventional formula (Collection / Total Assessment with Subsidy) * 100
        """
        if self.total_assessment_w_subsidy == 0:
            return 0.0
        self.collection_efficiency = (self.collection / self.total_assessment_w_subsidy) * 100
        return self.collection_efficiency

    def calculate_atc_loss(self):
        """
        Calculates AT&C Loss.
        Formula: T&D Loss + (100 - T&D Loss) * (100 - Collection Efficiency) / 100
        """
        td_loss = self.calculate_td_loss()
        collection_efficiency = self.calculate_collection_efficiency()
        self.atc_loss = td_loss + (100 - td_loss) * (100 - collection_efficiency) / 100
        return self.atc_loss
```

```python
metrics_results = []

for energy in input_energy_values:
    metrics = PowerUtilityMetrics(
        input_energy=energy,
        units_billed=units_billed,
        assessment_w_o_subsidy=assessment_w_o_subsidy,
        subsidy=subsidy,
        collection=collection
    )
    billing_efficiency = metrics.calculate_billing_efficiency()
    td_loss = metrics.calculate_td_loss()
    collection_efficiency = metrics.calculate_collection_efficiency()
    atc_loss = metrics.calculate_atc_loss()

    metrics_results.append({
        'input_energy': energy,
        'Billing Efficiency': billing_efficiency,
        'T&D Loss': td_loss,
        'Collection Efficiency': collection_efficiency,
        'AT&C Loss': atc_loss
    })

metrics_df = pd.DataFrame(metrics_results)
display(metrics_df.head())
```

```python
import pandas as pd
import numpy as np

# Simulate data for 100 hypothetical customers
num_customers = 100

# Simulate billing cycles (e.g., Monthly, Bi-monthly)
billing_cycles = np.random.choice(['Monthly', 'Bi-monthly'], size=num_customers, p=[0.8, 0.2])

# Create a DataFrame
billing_cycle_df = pd.DataFrame({'Customer ID': range(1, num_customers + 1),
                                 'Billing Cycle': billing_cycles})

print("Simulated Billing Cycle Data:")
display(billing_cycle_df.head())
```

### Simulate Payment Method Data

We can simulate different payment methods used by customers (e.g., Online, Check, Cash).

```python
# Simulate payment methods (e.g., Online, Check, Cash)
payment_methods = np.random.choice(['Online', 'Check', 'Cash'], size=num_customers, p=[0.6, 0.3, 0.1])

# Add to the DataFrame
billing_cycle_df['Payment Method'] = payment_methods

print("\nSimulated Billing Cycle and Payment Method Data:")
display(billing_cycle_df.head())
```

### Simulate Hypothetical Billing and Collection Data

We'll simulate a billed amount and a collected amount for each customer.

```python
# Simulate hypothetical billed and collected amounts for each customer
np.random.seed(42) # for reproducibility
billed_amounts = np.random.uniform(100, 500, size=num_customers)
# Simulate collection as a percentage of billed amount, with some variation
collection_rates = np.random.uniform(0.7, 1.0, size=num_customers)
collected_amounts = billed_amounts * collection_rates

# Create a DataFrame for billing and collection data
collection_data_df = pd.DataFrame({'Customer ID': range(1, num_customers + 1),
                                  'Billed Amount': billed_amounts,
                                  'Collected Amount': collected_amounts})

print("Simulated Billing and Collection Data:")
display(collection_data_df.head())
```

### Combine Data and Analyze Collection Efficiency by Billing Cycle

Now we'll combine the billing cycle data with the simulated billing and collection data and calculate the collection efficiency for each billing cycle.

```python
# Join the dataframes
merged_df = pd.merge(billing_cycle_df, collection_data_df, on='Customer ID')

# Group by Billing Cycle and calculate total billed and collected amounts
billing_cycle_collection = merged_df.groupby('Billing Cycle').agg(
    Total_Billed=('Billed Amount', 'sum'),
    Total_Collected=('Collected Amount', 'sum')
).reset_index()

# Calculate Collection Efficiency for each billing cycle
billing_cycle_collection['Collection Efficiency (%)'] = (billing_cycle_collection['Total_Collected'] / billing_cycle_collection['Total_Billed']) * 100

print("\nCollection Efficiency by Billing Cycle:")
display(billing_cycle_collection)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of Billed Amount vs. Collected Amount
plt.figure(figsize=(8, 6))
plt.scatter(merged_df['Billed Amount'], merged_df['Collected Amount'], alpha=0.5)
plt.title('Collected Amount vs. Billed Amount (Simulated Data)')
plt.xlabel('Billed Amount')
plt.ylabel('Collected Amount')
plt.grid(True)
plt.show()
```

### Analyze Collection Efficiency by Payment Method

We'll group the combined data by payment method and calculate the collection efficiency for each.

```python
# Group by Payment Method and calculate total billed and collected amounts
payment_method_collection = merged_df.groupby('Payment Method').agg(
    Total_Billed=('Billed Amount', 'sum'),
    Total_Collected=('Collected Amount', 'sum')
).reset_index()

# Calculate Collection Efficiency for each payment method
payment_method_collection['Collection Efficiency (%)'] = (payment_method_collection['Total_Collected'] / payment_method_collection['Total_Billed']) * 100

print("\nCollection Efficiency by Payment Method:")
display(payment_method_collection)
```
