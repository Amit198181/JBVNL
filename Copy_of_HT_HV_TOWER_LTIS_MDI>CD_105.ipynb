```python
import pandas as pd
from google.colab import files
import io

print("Please upload your Excel file:")
uploaded = files.upload()
```

```python
# Assuming only one file was uploaded
for filename in uploaded.keys():
    excel_file = io.BytesIO(uploaded[filename])
    df = pd.read_excel(excel_file)
    print(f"Successfully loaded '{filename}' into a pandas DataFrame.")
    display(df.head())
```

```python
print(f"DataFrame shape before removing duplicates: {df.shape}")
df.drop_duplicates(subset=['ACCOUNT_NO'], inplace=True)
print(f"DataFrame shape after removing duplicates by 'ACCOUNT_NO': {df.shape}")
```

```python
# Display all column names to identify relevant MDI and Billed Units columns
print(df.columns.tolist())
```

```python
# Define the MDI and Billed Units columns for the last four months
mdi_cols = ['MDI_HTS_202511', 'MDI_HTS_202512', 'MDI_HTS_202601', 'MDI_HTS_202602']
billed_units_cols = ['BILLED_UNITS_HTS_202511', 'BILLED_UNITS_HTS_202512', 'BILLED_UNITS_HTS_202601', 'BILLED_UNITS_HTS_202602']

# Filter for MDI = 0 for the last four months
meter_failure_mdi_filter = (df[mdi_cols] == 0).all(axis=1)

# Filter for monthly consumption > 100 kWh for the last four months
high_consumption_filter = (df[billed_units_cols] > 100).all(axis=1)

# Combine the filters
filtered_consumers = df[meter_failure_mdi_filter & high_consumption_filter]

# Get the list of Consumer IDs
consumer_ids = filtered_consumers['CONSNO'].tolist()

if consumer_ids:
    print("Consumer IDs matching the criteria (MDI=0 for last 4 months AND Billed Units > 100 kWh for last 4 months):\n")
    for cons_id in consumer_ids:
        print(cons_id)
    print("\nPotential causes for such a pattern (MDI=0 with high consumption) could include:")
    print("1. MDI sensor failure: The MDI meter component might be faulty and consistently reporting zero despite actual consumption.")
    print("2. Meter bypass or tampering: The meter's MDI recording mechanism might have been intentionally bypassed.")
    print("3. Data recording error: An error in the data collection or storage system leading to incorrect MDI readings.")
else:
    print("No consumers found matching the specified criteria.")
```

### Refined Analysis: MDI Threshold and Violation Frequency Across All Months

This section refines the analysis based on your request. We will calculate the `MDI_Threshold`, then determine `Violation_Frequency` by checking against *all available* monthly MDI columns, and finally summarize the high-violation consumers by division.

```python
import re

# 1. Ensure MDI_Threshold is calculated (re-calculate for clarity, though already done)
df['MDI_Threshold'] = df['CONTRACTED_LOAD'] * 1.05

# Identify all MDI columns dynamically from the DataFrame
mdi_columns_all = [col for col in df.columns if re.match(r'^MDI_HTS_\d{6}$', col)]

# 2. Perform row-wise count for Violation_Frequency
# Initialize a new column for violation counts with zeros
df['Violation_Frequency'] = 0

# For each identified MDI column, check if Actual MDI > MDI_Threshold and sum the occurrences
for col in mdi_columns_all:
    # Ensure the MDI column is numeric and handle potential non-numeric values by coercing to NaN
    # Then fill NaN with 0 for comparison if MDI is missing
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    df['Violation_Frequency'] += (df[col] > df['MDI_Threshold']).astype(int)

# 3. Filter the DataFrame for consumers where Violation_Frequency >= 4
high_violation_consumers = df[df['Violation_Frequency'] >= 4]

print(f"Number of consumers with Violation_Frequency >= 4: {len(high_violation_consumers)}")

# 4. Summarize the results by division
if not high_violation_consumers.empty:
    division_summary = high_violation_consumers.groupby('DIVISION').agg(
        num_consumers=('CONSNO', 'nunique'),
        avg_violation_frequency=('Violation_Frequency', 'mean'),
        max_violation_frequency=('Violation_Frequency', 'max')
    ).reset_index()

    print("\nSummary of High-Violation Consumers by Division:")
    display(division_summary.sort_values(by='num_consumers', ascending=False))
else:
    print("No consumers found with a Violation_Frequency of 4 or more.")
```

### Average of Top 4 MDI Values for High-Violation Consumers

```python
import numpy as np

# Identify all MDI columns dynamically
mdi_columns_all = [col for col in df.columns if re.match(r'^MDI_HTS_\d{6}$', col)]

# Function to calculate the average of the top N values
def average_top_n(row, columns, n=4):
    mdi_values = row[columns].dropna().sort_values(ascending=False)
    if len(mdi_values) >= n:
        return mdi_values.head(n).mean()
    elif len(mdi_values) > 0:
        return mdi_values.mean() # If fewer than N values, average available MDI values
    return np.nan # No MDI values available

# Apply the function to each row of high_violation_consumers using .loc to avoid SettingWithCopyWarning
high_violation_consumers.loc[:, 'Avg_Top_4_MDI'] = high_violation_consumers.apply(lambda row: average_top_n(row, mdi_columns_all, n=4), axis=1)

print("Calculated 'Avg_Top_4_MDI' for high-violation consumers.")
display(high_violation_consumers[['ACCOUNT_NO', 'CONTRACTED_LOAD', 'Avg_Top_4_MDI', 'Violation_Frequency']].head())
```
