# Task
Analyze the data in "TOWER_APR_AUG.xlsx" to determine if the maximum demand exceeded the contract demand by 5% or more at least four times during the last financial year.

## Load the data

### Subtask:
Load the data from the `TOWER_APR_AUG.xlsx` file into a pandas DataFrame.


**Reasoning**:
The subtask is to load the data from the Excel file into a pandas DataFrame and display the first 5 rows. This can be achieved by importing the pandas library, reading the Excel file using `pd.read_excel`, and displaying the head of the DataFrame.



```python
import pandas as pd

df = pd.read_excel('TOWER_APR_AUG.xlsx')
display(df.head())
```

## Inspect the data

### Subtask:
Display the first few rows and the columns of the DataFrame to understand its structure and identify the relevant columns for demand and contract demand.


**Reasoning**:
Display the first few rows, column information, and descriptive statistics of the DataFrame to understand its structure and content.



```python
display(df.head())
display(df.info())
display(df.describe())
```

## Calculate demand excess

### Subtask:
Create a new column to calculate the percentage by which the maximum demand exceeds the contract demand.


**Reasoning**:
Create new columns to calculate the percentage by which the maximum demand exceeds the contract demand for each month with MDI data.



```python
import numpy as np

mdi_columns = [col for col in df.columns if col.startswith('MDI_HTS_')]

for mdi_col in mdi_columns:
    month_year = mdi_col.split('_')[-1]
    excess_col_name = f'DEMAND_EXCESS_PERCENT_{month_year}'
    # Handle potential division by zero by replacing 0 in CONTRACTED_LOAD with NaN for calculation
    contracted_load_for_calc = df['CONTRACTED_LOAD'].replace(0, np.nan)
    df[excess_col_name] = ((df[mdi_col] - contracted_load_for_calc) / contracted_load_for_calc) * 100

display(df.filter(like='DEMAND_EXCESS_PERCENT_').head())
```

## Identify exceedance events

### Subtask:
Filter the data to identify instances where the demand excess is 5% or more.


**Reasoning**:
Create boolean columns indicating if the demand excess is 5% or more for each month, handling NaN values.



```python
demand_excess_cols = [col for col in df.columns if col.startswith('DEMAND_EXCESS_PERCENT_')]

for col in demand_excess_cols:
    exceeded_col_name = f'EXCEEDED_5_PERCENT_{col.split("_")[-1]}'
    # Create boolean column: True if demand excess is >= 5, False otherwise (including NaN)
    df[exceeded_col_name] = df[col] >= 5

display(df.filter(like='EXCEEDED_5_PERCENT_').head())
```

## Count exceedance events per financial year

### Subtask:
Count the number of times the exceedance condition was met within the last financial year based on the data available.


**Reasoning**:
Sum the boolean exceedance columns for each row and store the result in a new column. Then display the head of the dataframe including the new column.



```python
exceedance_cols = [col for col in df.columns if col.startswith('EXCEEDED_5_PERCENT_')]
df['EXCEEDANCE_COUNT_LAST_FY'] = df[exceedance_cols].sum(axis=1)
display(df[['EXCEEDANCE_COUNT_LAST_FY'] + exceedance_cols].head())
```

## Check the condition

### Subtask:
Determine if the count of exceedance events is four or more.


**Reasoning**:
Create a boolean variable to check if the count of exceedance events is four or more for at least one customer and print the result.



```python
exceeded_four_or_more_times = (df['EXCEEDANCE_COUNT_LAST_FY'] >= 4).any()
print(f"At least one customer exceeded the contract demand by 5% or more four or more times: {exceeded_four_or_more_times}")
```

## Report the result

### Subtask:
Output whether the condition (maximum demand exceeding contract demand by 5% or more at least four times) was met.


**Reasoning**:
Use a print statement to output whether the condition was met based on the value of the `exceeded_four_or_more_times` variable, as requested by the subtask.



```python
print(f"The condition (maximum demand exceeding contract demand by 5% or more at least four times) was met: {exceeded_four_or_more_times}")
```

## Summary:

### Data Analysis Key Findings

*   The analysis confirmed that there was at least one customer whose maximum demand exceeded their contract demand by 5% or more on four or more occasions within the available data for the last financial year.

# Task
Calculate the enhanced contract demand for each consumer based on the provided criteria and the uploaded data.

## Identify mdi columns

### Subtask:
Identify all columns in the DataFrame that represent maximum demand (MDI) readings.


**Reasoning**:
Identify the columns in the DataFrame that represent maximum demand (MDI) readings by iterating through the column names and checking if they start with 'MDI_HTS_'.



```python
mdi_columns = [col for col in df.columns if col.startswith('MDI_HTS_')]
print("MDI columns identified:", mdi_columns)
```

## Extract mdi values

### Subtask:
For each consumer, extract the MDI values from the identified columns.


**Reasoning**:
Extract the MDI values from the identified columns and display the first few rows.



```python
mdi_values = df[mdi_columns].values
display(mdi_values[:5])
```

## Find top four mdi values

### Subtask:
For each consumer, find the four highest MDI values from the extracted values.


**Reasoning**:
Sort the MDI values for each consumer in descending order and select the top four.



```python
top_four_mdi = np.sort(mdi_values, axis=1)[:, -4:][:, ::-1]
display(top_four_mdi[:5])
```

## Calculate the average of top four mdi values

### Subtask:
Calculate the average of the top four MDI values for each consumer.


**Reasoning**:
Calculate the mean of the top four MDI values for each consumer, handling NaN values, and store the result in a new variable, then display the first few values.



```python
average_top_four_mdi = np.nanmean(top_four_mdi, axis=1)
display(average_top_four_mdi[:5])
```

## Store enhanced contract demand

### Subtask:
Create a new column in the DataFrame to store the calculated enhanced contract demand for each consumer.


**Reasoning**:
Assign the calculated average top four MDI values to a new column in the DataFrame and display the head to verify.



```python
df['ENHANCED_CONTRACT_DEMAND'] = average_top_four_mdi
display(df[['ENHANCED_CONTRACT_DEMAND', 'CONTRACTED_LOAD']].head())
```

## Report enhanced contract demand

### Subtask:
Display the consumer account number and their calculated enhanced contract demand.


**Reasoning**:
Select and display the 'ACCOUNT_NO' and 'ENHANCED_CONTRACT_DEMAND' columns from the DataFrame.



```python
display(df[['ACCOUNT_NO', 'ENHANCED_CONTRACT_DEMAND']].head())
```
