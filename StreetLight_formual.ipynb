```python
import numpy as np

def extract_numeric(text):
    if isinstance(text, str):
        # Use regex to find all numbers (integers or floats) in the string
        numbers = re.findall(r'\d+\.?\d*', text.replace(',', ''))
        if numbers:
            # Try to convert the first found number to float
            try:
                return float(numbers[0])
            except ValueError:
                return np.nan
    return np.nan

import re

# Create copies of the columns to clean and convert
filtered_df_numeric = filtered_df.copy()

# Apply the extraction and conversion for 'Fixed/Demand Charges (Unit)'
filtered_df_numeric['Fixed/Demand Charges (Numeric)'] = filtered_df_numeric['Fixed/Demand Charges (Unit)'].apply(extract_numeric)

# Apply the extraction and conversion for 'Energy Charges (Paise/kWh, Rs/unit)'
filtered_df_numeric['Energy Charges (Numeric)'] = filtered_df_numeric['Energy Charges (Paise/kWh, Rs/unit)'].apply(extract_numeric)

# Filter rows where both extracted numeric columns are not NaN
numeric_data_df = filtered_df_numeric[
    filtered_df_numeric['Fixed/Demand Charges (Numeric)'].notna() &
    filtered_df_numeric['Energy Charges (Numeric)'].notna()
]

display(numeric_data_df)
```

```python
numeric_data_df = numeric_data_df[numeric_data_df['Tariff Scheme Name / Rate Code'] != 'Not Specified']
display(numeric_data_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Fixed/Demand Charges (Numeric)',
    y='Energy Charges (Numeric)',
    data=numeric_data_df,
    hue='State/UT', # Differentiate points by State/UT
    size='Consumer Category Name', # Use consumer category to vary point size
    sizes=(20, 200), # Range of point sizes
    alpha=0.7
)
plt.title('Fixed vs. Energy Charges (Numeric)')
plt.xlabel('Fixed/Demand Charges (Numeric)')
plt.ylabel('Energy Charges (Numeric)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) # Move legend outside to prevent overlap
plt.tight_layout() # Adjust layout to prevent labels/legend from being cut off
plt.show()
```
