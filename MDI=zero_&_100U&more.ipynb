```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # Read the CSV into a pandas DataFrame
  try:
    df = pd.read_excel(fn)
    print('\nDataFrame head:')
    display(df.head())
  except Exception as e:
    print('Error reading file: {e}'.format(e=e))
```

MDI= 0 for all the 4 M

```python
# Select the Billed Units columns for the four months
billed_units_columns = ['BILLED_UNITS_HTS_202505', 'BILLED_UNITS_HTS_202506', 'BILLED_UNITS_HTS_202507', 'BILLED_UNITS_HTS_202508']
mdi_columns = ['MDI_HTS_202505', 'MDI_HTS_202506', 'MDI_HTS_202507', 'MDI_HTS_202508']

# Filter the DataFrame where all MDI columns are 0 and the last month's Billed Units are 100 or above
consumption_mdi_filtered_df = df[(df[mdi_columns] == 0).all(axis=1) & (df['BILLED_UNITS_HTS_202508'] >= 100)]

# Display the filtered DataFrame
display(consumption_mdi_filtered_df)
```

```python
consumption_filtered_df.to_excel('MDI_ZERO_100U.xlsx', index=False)
files.download('MDI_ZERO_100U.xlsx')
```
