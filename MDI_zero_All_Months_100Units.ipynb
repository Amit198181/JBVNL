# Task
Analyze the dataframe `df` to find rows where the maximum demand (MDI) is zero for all months.

## Identify mdi columns

### Subtask:
Find all columns in the DataFrame that represent the maximum demand (MDI) for each month.


**Reasoning**:
Identify all columns in the DataFrame that contain 'MDI' in their name.



```python
mdi_columns = [col for col in df.columns if 'MDI' in col]
print("MDI columns:", mdi_columns)
```

## Check for zero mdi across months

### Subtask:
Create a condition to check if the value in all identified MDI columns is zero for each row.


**Reasoning**:
Create a boolean condition to check if all values in the identified MDI columns are zero for each row, handling potential NaN values by treating them as non-zero for this condition.



```python
zero_mdi_condition = (df[mdi_columns].fillna(-1) == 0).all(axis=1)
```

## Filter dataframe

### Subtask:
Apply the condition to filter the DataFrame and keep only the rows where MDI is zero across all months.


**Reasoning**:
Apply the previously created boolean condition `zero_mdi_condition` to filter the DataFrame `df` and store the result in a new DataFrame `df_zero_mdi`.



```python
df_zero_mdi = df[zero_mdi_condition]
```

## Display results

### Subtask:
Show the filtered DataFrame to the user.


**Reasoning**:
Display the first few rows of the filtered DataFrame and print the number of rows.



```python
display(df_zero_mdi.head())
print(f"Number of rows with zero MDI across all months: {len(df_zero_mdi)}")
```
