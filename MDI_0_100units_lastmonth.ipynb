# Task
Analyze the uploaded Excel file to identify consumers who had a maximum demand of zero in August and consumed 100 units or more in the same month. Display the segregated data.

## Load the data

### Subtask:
Load the uploaded Excel file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the Excel file into a DataFrame, then display the head and info of the DataFrame to understand its structure.



```python
import pandas as pd

excel_file = list(uploaded.keys())[0]
df = pd.read_excel(excel_file)

display(df.head())
display(df.info())
```

## Filter for august data with zero maximum demand

### Subtask:
Filter the DataFrame to include only records from August where the maximum demand is zero.


**Reasoning**:
Filter the DataFrame to include only records from August where the maximum demand is zero, based on the identified column `MDI_HTS_202508`.



```python
df_aug_zero_demand = df[df['MDI_HTS_202508'] == 0]
display(df_aug_zero_demand.head())
```

## Segregate consumers with 100 units or more

### Subtask:
From the filtered data (`df_aug_zero_demand`), identify the consumers with a consumption of 100 units or more.


**Reasoning**:
Filter the DataFrame `df_aug_zero_demand` to find consumers with billed units greater than or equal to 100 in August 2025.



```python
df_aug_high_consumption = df_aug_zero_demand[df_aug_zero_demand['BILLED_UNITS_HTS_202508'] >= 100]
display(df_aug_high_consumption.head())
```

## Present the results

### Subtask:
Display the segregated data showing the consumers who meet both conditions.


**Reasoning**:
Display the DataFrame containing consumers with zero maximum demand and high consumption in August.



```python
display(df_aug_high_consumption)
```

## Summary:

### Data Analysis Key Findings

*   The analysis identified consumers who had a maximum demand of zero in August 2025 and consumed 100 units or more in the same month.
*   The data for these consumers is stored in the `df_aug_high_consumption` DataFrame.

### Insights or Next Steps

## Download the segregated data

### Subtask:
Download the segregated DataFrame (`df_aug_high_consumption`) as an Excel file.

**Reasoning**:
Use the `to_excel` method to convert the DataFrame to an Excel file and then use `files.download` to download the file.

```python
output_filename = 'consumers_zero_demand_high_consumption_august.xlsx'
df_aug_high_consumption.to_excel(output_filename, index=False)

files.download(output_filename)
```
