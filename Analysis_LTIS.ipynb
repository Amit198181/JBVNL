## Analyze the data

### Subtask:
Analyze the data in the 'OVER_DRAWL_wrt_CD' sheet to understand the patterns and characteristics of consumers who are overdrawing their contracted load.


**Reasoning**:
Analyze the distribution of 'Total Overdrawn Count', the relationship between 'CONTRACTED_LOAD' and 'Total Overdrawn Count', investigate which geographical areas have the highest occurrences of overdrawn consumers, examine the 'Over_all_result' column, and identify the correlation between 'CONTRACTED_LOAD' and 'Over_all_result'.



```python
selected_sheet_name = 'OVER_DRAWL_wrt_CD'
df_selected_sheet = all_sheets[selected_sheet_name]

# 1. Analyze the distribution of 'Total Overdrawn Count'
print("Distribution of 'Total Overdrawn Count':")
print(df_selected_sheet['Total Overdrawn Count'].value_counts().sort_index())

# 2. Analyze the relationship between 'CONTRACTED_LOAD' and 'Total Overdrawn Count'
print("\nAverage 'CONTRACTED_LOAD' for each 'Total Overdrawn Count' group:")
print(df_selected_sheet.groupby('Total Overdrawn Count')['CONTRACTED_LOAD'].mean())

# 3. Investigate which Zones, Circles, and Divisions have the highest occurrences of overdrawn consumers
print("\nTop 10 Zones with highest 'Total Overdrawn Count':")
print(df_selected_sheet.groupby('ZONE')['Total Overdrawn Count'].sum().sort_values(ascending=False).head(10))

print("\nTop 10 Circles with highest 'Total Overdrawn Count':")
print(df_selected_sheet.groupby('CIRCLE')['Total Overdrawn Count'].sum().sort_values(ascending=False).head(10))

print("\nTop 10 Divisions with highest 'Total Overdrawn Count':")
print(df_selected_sheet.groupby('DIVISION')['Total Overdrawn Count'].sum().sort_values(ascending=False).head(10))

# 4. Examine the 'Over_all_result' column
print("\nPercentage of consumers with 'Over_all_result' as True:")
percentage_overall_true = (df_selected_sheet['Over_all_result'].sum() / len(df_selected_sheet)) * 100
print(f"{percentage_overall_true:.2f}%")

# 5. Identify correlation between 'CONTRACTED_LOAD' and 'Over_all_result'
print("\nMean 'CONTRACTED_LOAD' for 'Over_all_result' groups:")
print(df_selected_sheet.groupby('Over_all_result')['CONTRACTED_LOAD'].mean())

# Calculate and print the requested metrics for this sheet
print("\n--- Metrics for OVER_DRAWL_wrt_CD ---")
average_contracted_load = df_selected_sheet['CONTRACTED_LOAD'].mean()
print(f"Average contracted load: {average_contracted_load:.2f}")

contracted_load_range = f"{df_selected_sheet['CONTRACTED_LOAD'].min():.2f} - {df_selected_sheet['CONTRACTED_LOAD'].max():.2f}"
print(f"Contracted load range: {contracted_load_range}")

# Assuming MDI columns are those containing 'MDI_HTS_'
mdi_cols = [col for col in df_selected_sheet.columns if 'MDI_HTS_' in col]
average_mdi = df_selected_sheet[mdi_cols].mean().mean() # Average of averages across MDI columns
print(f"Average MDI (Maximum Demand Indicator): {average_mdi:.2f}")

# Maximum MDI value across all MDI columns and all consumers
max_mdi = df_selected_sheet[mdi_cols].max().max()
print(f"Maximum demand (MDI value): {max_mdi:.2f}")

total_consumers = len(df_selected_sheet)
print(f"Total Number of Consumers: {total_consumers}")

total_overdrawal_occurrences = df_selected_sheet['Total Overdrawn Count'].sum()
print(f"Total overdrawal occurrences (all months, all customers): {total_overdrawal_occurrences}")
```
