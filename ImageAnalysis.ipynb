````python
import pandas as pd
import os

# Path to the /content/ directory
content_path = '/content/'

# Get all items in /content/ and filter for Excel files
all_content_items = os.listdir(content_path)
excel_files_in_content = [item for item in all_content_items if item.endswith(('.xlsx', '.xls'))]

# Ensure loaded_dfs is initialized if it doesn't exist
if 'loaded_dfs' not in locals() and 'loaded_dfs' not in globals():
    loaded_dfs = {}

if not excel_files_in_content:
    print(f"No Excel files found in '{content_path}'.")
else:
    print(f"Found {len(excel_files_in_content)} Excel file(s) in '{content_path}'. Loading files...")
    for file_name in excel_files_in_content:
        full_path = os.path.join(content_path, file_name)
        try:
            df_name = os.path.splitext(file_name)[0] # Use file name without extension as key

            # Ensure unique names if files with same names exist (simple approach for now)
            original_df_name = df_name
            counter = 1
            while df_name in loaded_dfs:
                df_name = f"{original_df_name}_{counter}"
                counter += 1

            loaded_dfs[df_name] = pd.read_excel(full_path)
            print(f"\nSuccessfully loaded '{file_name}' as '{df_name}'. First 5 rows:")
            display(loaded_dfs[df_name].head())
        except Exception as e:
            print(f"Error loading '{file_name}': {e}")

    if any(df_name in loaded_dfs for df_name in [os.path.splitext(f)[0] for f in excel_files_in_content]):
        print(f"\nSuccessfully loaded new DataFrames from '{content_path}' into 'loaded_dfs' dictionary.")
        print(f"Total DataFrames now in 'loaded_dfs': {len(loaded_dfs)}")
    else:
        print("No new DataFrames were successfully loaded from '/content/' folder.")
````
