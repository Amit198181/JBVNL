## Visualize comparison

**Reasoning**:

```python
import matplotlib.pyplot as plt
import seaborn as sns

model_projected_df = model_projected_df.reset_index()
melted_df = model_projected_df.melt(id_vars='index', var_name='Fiscal Year', value_name='Projected Sales')
melted_df = melted_df.rename(columns={'index': 'Model'})

plt.figure(figsize=(10, 6))
sns.lineplot(data=melted_df, x='Fiscal Year', y='Projected Sales', hue='Model')
plt.title("Projected Sales Comparison of Models (FY26-27 to FY30-31)")
plt.xlabel("Fiscal Year")
plt.ylabel("Projected Sales (MU)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Data preparation

### Subtask:
Prepare the data for analysis by handling missing values and cleaning column names.


**Reasoning**:
Inspect the dataframe for missing values and handle them by imputing with the median, as the describe output showed some outliers in the 'Per Unit Rate (Rs per unit)' column. Also, clean the column name 'Per Unit Rate\n(Rs per unit)' by removing the newline character and any leading/trailing whitespace. Finally, display the first few rows and information of the cleaned dataframe to verify the changes.



```python
display(df.info())

# Handle missing values in 'Allocation in MU' and 'Per Unit Rate\n(Rs per unit)'
for col in ['Allocation in MU', 'Per Unit Rate\n(Rs per unit)']:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Clean the column name 'Per Unit Rate\n(Rs per unit)'
df.rename(columns={'Per Unit Rate\n(Rs per unit)': 'Per Unit Rate (Rs per unit)'}, inplace=True)

display(df.head())
display(df.info())
```

## Exploratory data analysis (eda)

### Subtask:
Analyze the relationships between potential factors ('Allocation in MU' and 'Per Unit Rate (Rs per unit)') and projected sales ('Projected Quantum purchase (MU)' columns) using visualizations and statistical methods.


**Reasoning**:
Create scatter plots to visualize the relationship between 'Allocation in MU' and each of the 'Projected Quantum purchase (MU)' columns and between 'Per Unit Rate (Rs per unit)' and each of the 'Projected Quantum purchase (MU)' columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

projected_columns = [col for col in df.columns if 'Projected Quantum purchase (MU)' in col]

# Scatter plots for 'Allocation in MU' vs Projected Sales
plt.figure(figsize=(15, 10))
for i, col in enumerate(projected_columns):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(data=df, x='Allocation in MU', y=col)
    plt.title(f'Allocation in MU vs {col}')
    plt.xlabel('Allocation in MU')
    plt.ylabel('Projected Sales (MU)')
plt.tight_layout()
plt.show()

# Scatter plots for 'Per Unit Rate (Rs per unit)' vs Projected Sales
plt.figure(figsize=(15, 10))
for i, col in enumerate(projected_columns):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(data=df, x='Per Unit Rate (Rs per unit)', y=col)
    plt.title(f'Per Unit Rate vs {col}')
    plt.xlabel('Per Unit Rate (Rs per unit)')
    plt.ylabel('Projected Sales (MU)')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Calculate and display the correlation matrix for 'Allocation in MU', 'Per Unit Rate (Rs per unit)', and all 'Projected Quantum purchase (MU)' columns.



```python
correlation_columns = ['Allocation in MU', 'Per Unit Rate (Rs per unit)'] + projected_columns
correlation_matrix = df[correlation_columns].corr()
display(correlation_matrix)
```

## Statistical modeling

### Subtask:
Build statistical models (e.g., regression) to quantify the impact of identified factors on projected sales.


**Reasoning**:
Build a linear regression model with 'Allocation in MU' as the independent variable and a projected sales column as the dependent variable, then repeat the process with 'Per Unit Rate (Rs per unit)' as the independent variable. Print the intercept and coefficient for both models.



```python
from sklearn.linear_model import LinearRegression

# Model 1: Allocation in MU vs Projected Sales (using FY26-27)
X1 = df['Allocation in MU'].values.reshape(-1, 1)
y = df['Projected Quantum purchase (MU) for FY26-27']

model1 = LinearRegression()
model1.fit(X1, y)

print("Model 1: Allocation in MU vs Projected Sales (FY26-27)")
print(f"Intercept: {model1.intercept_}")
print(f"Coefficient: {model1.coef_[0]}")
print("-" * 30)

# Model 2: Per Unit Rate (Rs per unit) vs Projected Sales (using FY26-27)
X2 = df['Per Unit Rate (Rs per unit)'].values.reshape(-1, 1)

model2 = LinearRegression()
model2.fit(X2, y)

print("Model 2: Per Unit Rate (Rs per unit) vs Projected Sales (FY26-27)")
print(f"Intercept: {model2.intercept_}")
print(f"Coefficient: {model2.coef_[0]}")
print("-" * 30)

print("Implications of high correlation between 'Allocation in MU' and 'Per Unit Rate (Rs per unit)':")
print("If both 'Allocation in MU' and 'Per Unit Rate (Rs per unit)' were included in the same linear regression model, multicollinearity would be a concern. This could lead to unstable coefficient estimates, making it difficult to interpret the individual impact of each variable on projected sales.")
```
