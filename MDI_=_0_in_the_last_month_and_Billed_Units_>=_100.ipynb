```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # Read the CSV into a pandas DataFrame
  try:
    df = pd.read_excel(fn)
    print('\nDataFrame head:')
    display(df.head())
  except Exception as e:
    print('Error reading file: {e}'.format(e=e))
```

```python
# Identify the last MDI column
mdi_columns = [col for col in df.columns if 'MDI' in col]
last_mdi_column = mdi_columns[-1]

# Identify the last BILLED_UNITS column
billed_units_columns = [col for col in df.columns if 'BILLED_UNITS' in col]
last_billed_units_column = billed_units_columns[-1]

# Filter rows based on the two conditions
filtered_df = df[(df[last_mdi_column] == 0) & (df[last_billed_units_column] >= 100)]

print("\nDataFrame filtered for MDI = 0 in the last month and Billed Units >= 100:")
display(filtered_df.head())
```

```python
from google.colab import files

# Save the filtered DataFrame to an Excel file
filtered_df.to_excel("MDI=0_100UNITS_lastMonth.xlsx", index=False)

# Download the file
files.download("MDI=0_100UNITS_lastMonth.xlsx")
```
