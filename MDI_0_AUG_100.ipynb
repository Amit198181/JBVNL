# Task
Find out the consumers having the maximum demand of zero in the month of August, with 100 units and more consumption of electricity from the file "LTIS_APR_AUG.xlsx".

## Load data

### Subtask:
Load the data from the uploaded Excel file "LTIS_APR_AUG.xlsx" into a pandas DataFrame.


**Reasoning**:
Import pandas and load the Excel file into a DataFrame, then display the head and info.



```python
import pandas as pd

df = pd.read_excel('LTIS_APR_AUG.xlsx')
display(df.head())
display(df.info())
```

## Filter for august data

### Subtask:
Filter the DataFrame to include only the data for the month of August.


**Reasoning**:
Filter the DataFrame to include only the data for the month of August by selecting rows where the `BILLED_UNITS_HTS_202508` column is not null.



```python
df_august = df[df['BILLED_UNITS_HTS_202508'].notnull()].copy()
display(df_august.head())
display(df_august.info())
```

## Filter for maximum demand of zero

### Subtask:
Filter the August data to include only consumers with a maximum demand of zero.


**Reasoning**:
Filter the `df_august` DataFrame to select rows where the 'MDI_HTS_202508' column is equal to 0 and display the head and information of the resulting DataFrame.



```python
df_august_zero_demand = df_august[df_august['MDI_HTS_202508'] == 0].copy()
display(df_august_zero_demand.head())
display(df_august_zero_demand.info())
```

## Filter for consumption of 100 units or more

### Subtask:
From the filtered data, identify consumers with 100 units or more consumption.


**Reasoning**:
Filter the DataFrame to include only consumers with 100 units or more consumption.



```python
df_august_high_consumption = df_august_zero_demand[df_august_zero_demand['BILLED_UNITS_HTS_202508'] >= 100].copy()
display(df_august_high_consumption.head())
display(df_august_high_consumption.info())
```

**Reasoning**:
Extract the unique consumer identifiers from the final filtered data and display the list of unique consumers.



```python
unique_consumers = df_august_high_consumption['CONSNO'].unique()
print("Unique consumers with zero maximum demand and 100 or more units consumption in August:")
print(unique_consumers)
```

## Identify unique consumers

### Subtask:
Extract the unique consumer identifiers from the final filtered data.


**Reasoning**:
Extract the unique consumer identifiers from the 'CONSNO' column of the filtered data and store them in a variable.



```python
unique_consumers = df_august_high_consumption['CONSNO'].unique()
print("Unique consumers with zero maximum demand and 100 or more units consumption in August:")
print(unique_consumers)
```

## Display results

### Subtask:
Display the list of consumers who meet all the criteria.


**Reasoning**:
Display the list of unique consumers who meet all the criteria.



```python
print("List of unique consumers with zero maximum demand and 100 or more units consumption in August:")
for consumer in unique_consumers:
    print(consumer)
```
