```python
# Identify the columns related to Maximum Demand (assuming they start with 'MDI_HTS_')
max_demand_cols = [col for col in df.columns if col.startswith('MDI_HTS_')]

# Filter data where maximum demand for all identified months is zero
# We will also filter for Consumption >= 100, using 'BILLED_UNITS_HTS_202508' as an example consumption column
filtered_df = df[(df[max_demand_cols] == 0).all(axis=1)]

# Now, filter the already filtered data for consumption in August (BILLED_UNITS_HTS_202508) >= 100
consumption_august_100_plus_df = filtered_df[filtered_df['BILLED_UNITS_HTS_202508'] >= 100]


# Display the first 5 rows of the resulting data
display(consumption_august_100_plus_df.head())
```
