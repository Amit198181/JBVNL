```python
# Helper function to extract a value by variable name
def get_value(var_str, df):
    # Use the actual column names instead of numerical indices
    row = df[df['Particular'].str.contains(var_str, case=False, na=False)]
    return float(row['DS-1 (A)'].iloc[0]) if not row.empty else None
```

```python
# Extract required variables
sale_of_energy = get_value('Total  Sale of Energy', df)
total_other_charges = get_value('Total Other Charges', df)
electricity_duty = get_value('Electricity duty', df)
subsidy = get_value('subsidy', df)
energy_sold = get_value('Total Quantity of Energy Sold', df)
```

```python
# Compute ratios and metrics
revenue_per_unit = sale_of_energy / energy_sold if sale_of_energy and energy_sold else None
other_receipts_per_unit = total_other_charges / energy_sold if total_other_charges and energy_sold else None
electricity_duty_per_unit = electricity_duty / energy_sold if electricity_duty and energy_sold else None
subsidy_per_unit = subsidy / energy_sold if subsidy and energy_sold else None

total_receipts = sum([x for x in [sale_of_energy, total_other_charges, electricity_duty, subsidy] if x])
revenue_receipts_ratio = sale_of_energy / total_receipts if sale_of_energy and total_receipts else None
other_receipts_ratio = total_other_charges / total_receipts if total_other_charges and total_receipts else None
ed_receipts_ratio = electricity_duty / total_receipts if electricity_duty and total_receipts else None
subsidy_receipts_ratio = subsidy / total_receipts if subsidy and total_receipts else None
```

```python
# Print all results
print(f'Revenue per unit sold: {revenue_per_unit:.2f}')
print(f'Other Receipts per unit sold: {other_receipts_per_unit:.2f}')
print(f'Electricity duty per unit sold: {electricity_duty_per_unit:.2f}')
print(f'Subsidy per unit sold: {subsidy_per_unit:.2f}')
print(f'Revenue-to-total Receipts Ratio: {revenue_receipts_ratio:.2%}')
print(f'Other Receipts-to-total Receipts Ratio: {other_receipts_ratio:.2%}')
print(f'Electricity Duty-to-total Receipts Ratio: {ed_receipts_ratio:.2%}')
print(f'Subsidy-to-total Receipts Ratio: {subsidy_receipts_ratio:.2%}')
```
