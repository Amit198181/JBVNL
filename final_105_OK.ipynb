# Task
Analyze the dataframe to identify consumers where the maximum demand was recorded to be in excess of the contract demand by 5% or more at least four times during the last financial year.

## Identify mdi columns

### Subtask:
Identify the columns that contain MDI data for each month.


**Reasoning**:
Identify the columns that contain MDI data for each month by iterating through the column names and checking if they contain 'MDI_HTS_'.



```python
mdi_columns = [col for col in df.columns if 'MDI_HTS_' in col]
display(mdi_columns)
```

## Calculate excess demand

### Subtask:
For each consumer and each month, calculate if the MDI exceeds the `CONTRACTED_LOAD` by 5% or more.


**Reasoning**:
Calculate for each month if the MDI exceeds the contracted load by 5% or more and store the results in new columns.



```python
for mdi_col in mdi_columns:
    excess_col_name = mdi_col + '_excess'
    df[excess_col_name] = df[mdi_col] >= (df['CONTRACTED_LOAD'] * 1.05)

display(df.head())
```

## Count occurrences of excess demand

### Subtask:
For each consumer, count the number of times the excess demand condition was met across all months.


**Reasoning**:
Select the columns ending with '_excess', sum the boolean values across these columns for each row, store the sum in a new column 'excess_demand_count', and display the head of the dataframe.



```python
excess_columns = [col for col in df.columns if col.endswith('_excess')]
df['excess_demand_count'] = df[excess_columns].sum(axis=1)
display(df.head())
```

## Filter consumers

### Subtask:
Filter the dataframe to include only consumers where the excess demand condition was met at least four times.


**Reasoning**:
Filter the dataframe to keep only rows where the excess_demand_count is 4 or more.



```python
high_demand_consumers = df[df['excess_demand_count'] >= 4]
display(high_demand_consumers.head())
```

## Display results

### Subtask:
Display the information for the identified consumers.


**Reasoning**:
Display the high_demand_consumers dataframe to show the information for the identified consumers.



```python
display(high_demand_consumers)
```

*   New boolean columns were successfully added to the dataframe, indicating whether the MDI for each month exceeded 105% of the `CONTRACTED_LOAD` for each consumer.

# Task
For the high-demand consumers identified in the previous step, calculate the enhanced contract demand as the average of the four highest MDI readings from the last financial year and add this as a new column to the dataframe.

## Select mdi columns

### Subtask:
Select only the columns that contain MDI data.


**Reasoning**:
Select only the columns that contain MDI data from the high_demand_consumers dataframe.



```python
mdi_data_high_demand = high_demand_consumers[mdi_columns]
display(mdi_data_high_demand.head())
```

## Find the four highest mdi values

### Subtask:
For each high-demand consumer, find the four highest MDI values across the selected MDI columns.


**Reasoning**:
Sort the MDI values for each consumer in descending order and select the top four to find the four highest MDI values.



```python
top_four_mdi = mdi_data_high_demand.apply(lambda row: row.sort_values(ascending=False).head(4), axis=1)
display(top_four_mdi.head())
```

## Calculate the average of the four highest mdi values

### Subtask:
For each high-demand consumer, calculate the average of the four highest MDI values.


**Reasoning**:
Calculate the average of the four highest MDI values for each consumer.



```python
enhanced_contract_demand_values = top_four_mdi.mean(axis=1)
display(enhanced_contract_demand_values.head())
```

## Add enhanced contract demand to the dataframe

### Subtask:
Add a new column to the `high_demand_consumers` dataframe containing the calculated enhanced contract demand.


**Reasoning**:
Assign the calculated enhanced contract demand values to a new column in the high_demand_consumers dataframe and display the head of the dataframe to verify the addition.



```python
high_demand_consumers['enhanced_contract_demand'] = enhanced_contract_demand_values
display(high_demand_consumers.head())
```

## Display results

### Subtask:
Display the `high_demand_consumers` dataframe with the new 'enhanced_contract_demand' column.


**Reasoning**:
Display the high_demand_consumers dataframe with the new 'enhanced_contract_demand' column.



```python
display(high_demand_consumers)
```
